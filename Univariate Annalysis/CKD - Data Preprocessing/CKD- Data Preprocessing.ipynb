```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if (dataset[columnName].dtype=='O'):
           #print("qual")
           qual.append(columnName)
        else:
           #print("quan")
           quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(dataset)
```

```python
num_columns = ['age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo']
```

```python
from sklearn.preprocessing import StandardScaler
num_columns = ['age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo']
for columns in num_columns:
    dataset[columns].fillna(dataset[columns].median(), inplace=True)

scaler = StandardScaler()
dataset[num_columns] = scaler.fit_transform(dataset[num_columns])
```

```python
cat_columns = ['pcc', 'ba', 'pcv', 'wc', 'dm', 'cad', 'appet', 'pe', 'ane']
```
