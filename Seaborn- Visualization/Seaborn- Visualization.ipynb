## Univariate Analysis

```python
import seaborn as sns
```

```python
sns.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

## Bivariate Analysis : Distribution Plot

```python
sns.jointplot(x='ssc_p',y='salary',data=dataset)
plt.show()
```

```python
sns.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
sns.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
```

```python
# Pair plot
```

```python
sns.pairplot(dataset,hue='gender',diag_kind='kde',kind='scatter',palette='husl')
plt.show()
```

## For Categorical plot

```python
sns.stripplot(x='ssc_b',y='ssc_p',data=dataset)
plt.show()
```

```python
sns.stripplot(x='gender',y='ssc_p',data=dataset,jitter=True)
plt.show()
```

```python
sns.swarmplot(x='gender',y='ssc_p',data=dataset)
plt.show()
```

## Distribution Observation

```python
sns.violinplot(x='gender',y='ssc_p',hue='gender',data=dataset)
plt.show()
```

```python
sns.violinplot(x='specialisation',y='salary',hue='gender',data=dataset)
plt.show()
```

## Box Plot

```python
sns.boxplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

## Bar Plot

```python
sns.barplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

## Count Plot

```python
sns.countplot(x='ssc_b',data=dataset,hue='gender')
plt.show()
```

## Point Plot

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df=sns.load_dataset('exercise')
sns.catplot(x='time',y='pulse',hue='kind',kind='violin',col='diet',data=df)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df=sns.load_dataset('tips')
sns.regplot(x='total_bill',y='tip',data=df)
sns.lmplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df=sns.load_dataset('anscombe')
sns.lmplot(x='x',y='y',data=df.query("dataset=='II'"))
plt.show()
           
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df=sns.load_dataset('anscombe')
sns.lmplot(x='x',y='y',data=df.query("dataset=='II'"),order=2)
plt.show()
```
